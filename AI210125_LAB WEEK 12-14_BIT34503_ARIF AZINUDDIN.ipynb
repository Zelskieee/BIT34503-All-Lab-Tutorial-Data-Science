```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from IPython.display import Markdown

message = "<b>Importing the dataset</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from IPython.display import Markdown

message = "<b>Splitting the dataset into the Training set and Test set</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
from IPython.display import Markdown

message = "<b>Feature Scaling</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from IPython.display import Markdown

message = "<b>Training the Logistic Regression model on the Training set</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
from IPython.display import Markdown

message = "<b>Predicting a New Results</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
from IPython.display import Markdown

message = "<b>Predicting the Test Set Results</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from IPython.display import Markdown

message = "<b>Making the Confusion Matrix</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from IPython.display import Markdown

message = "<b>Visualizing the Training Set Results</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=0.25),
                     np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=0.25))

cmap = ListedColormap(['#FFA07A', '#1E90FF'])  

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha=0.75, cmap=cmap)

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c=cmap(i), label=j)

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=0.25),
                     np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=0.25))

cmap = ListedColormap(['#FFA07A', '#1E90FF'])  

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha=0.75, cmap=cmap)

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c=cmap(i), label=j)

plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
