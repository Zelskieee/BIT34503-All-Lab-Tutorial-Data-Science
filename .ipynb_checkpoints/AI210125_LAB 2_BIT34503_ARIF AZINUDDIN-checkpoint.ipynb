```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
#exporting dataset
url= "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
ds = pd.read_csv(url, header = None)
```

```python
# create headers List
header=["symboling", "normalized-losses","make","fuel-type",
"aspiration","num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base", "length","width","height",
"curb-weight", "engine-type","num-of-cylinders", "engine-size",
"fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
```

```python
ds.columns=header
```

```python
# replace "?" to NaN
ds.replace("?", np.nan, inplace=True)
ds
```

```python
missing_data = ds.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist(): 
    print(column)
    print (missing_data[column].value_counts()) 
    print("")
```

```python
ds['bore'] = ds['bore'].fillna(0)
ds['horsepower'] = ds['horsepower'].fillna(0) 
ds['peak-rpm'] = ds['peak-rpm'].fillna(0)
```

```python
print(ds)
```

```python
#replace the missing 'num-of-doors' values by the most frequent
ds['num-of-doors'].value_counts()
ds['num-of-doors'].value_counts().idxmax()
ds["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
avg_norm_loss = ds["normalized-losses"].astype("float").mean(axis=0) 
print("Average of normalized-losses: ", avg_norm_loss)
```

```python
ds["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore=ds['bore'].astype('float').mean(axis=0) 
print("Average of bore:", avg_bore)
```

```python
# simply drop whole row with NaN in "price" column 
ds.dropna (subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows 
ds.reset_index(drop=True, inplace=True)
```

```python
#Convert data types to proper format
ds["bore"] = ds["bore"].astype(float)
ds["stroke"] = ds["stroke"].astype(float)
ds["normalized-losses"] = ds["normalized-losses"].astype(int)
ds["price"] = ds["price"].astype(float)
ds["peak-rpm"] = ds["peak-rpm"].astype(float)
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg) 
ds['city-L/100km'] = 235/ds["city-mpg"]

# check your transformed data
ds.head()
```

```python
# Convert mpg to L/100km and rename the column
ds["highway-L/100km"] = 235 / ds["highway-mpg"]

# Drop the original "highway-mpg" column if needed
# ds.drop("highway-mpg", axis=1, inplace=True)

# check your transformed data
ds.head()
```

```python
# replace (original value) by (original value)/(maximum value) 
ds['length'] = ds['length']/ds['length'].max()
ds['width'] = ds['width']/ds['width'].max()
```

```python
# Normalize the 'height' column
ds['height'] = ds['height'] / ds['height'].max()
```

```python
print(ds[['length', 'width', 'height']])
```
