```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.svm import SVC  # Import the Support Vector Classifier (SVC)

# Initialize the SVM classifier
classifier = SVC(kernel='rbf', random_state=0)  # 'rbf' kernel is used here; you can explore other kernels as well

# Fit the SVM classifier to the training data
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from IPython.display import Markdown

message = "<b>Support Vector Machine for Training Set</b>"
Markdown(f"<div style='font-size: xx-large; font-family: Arial, sans-serif;'>{message}</div>")
```

```python
from matplotlib.colors import ListedColormap

# No need to inverse transform for SVM as scaling isn't required
X_set, y_set = X_train, y_train

# Create a meshgrid for plotting the decision boundaries
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

# Use the SVM classifier to predict on the meshgrid points
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plotting the actual training data points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
                c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('Support Vector Machines (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
